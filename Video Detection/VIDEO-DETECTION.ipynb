```python
# N√≥dulo 1
# Instalando as bibliotecas necess√°rias

!pip install imageio
!pip install imageio[ffmpeg]
!pip install imageio-ffmpeg
import imageio

import imageio
from tkinter import Tk, Label
from PIL import Image, ImageTk
import threading
 
def exibir_video_vscode(video_caminho):
    # Carrega o v√≠deo usando o backend ffmpeg
    video = imageio.get_reader(video_caminho, 'ffmpeg')
    root = Tk()
    lbl = Label(root)
    lbl.pack()
 
    def stream():
        try:
            for frame in video.iter_data():
                img = ImageTk.PhotoImage(Image.fromarray(frame))
                lbl.config(image=img)
                lbl.image = img  # Atualiza a imagem no widget
                root.update()
        except RuntimeError:
            pass
 
    # Rodar o v√≠deo em uma thread separada para evitar congelamento
    threading.Thread(target=stream).start()
    root.mainloop()
 
```

```python
# N√≥dulo 2
exibir_video_vscode('runs/detect/predict/video1.mp4')
```

```python
# N√≥dulo 3
exibir_video_vscode('runs/detect/predict2/video1.mp4')
```

```python
# N√≥dulo 1
def exibir_video_jupyter(video_caminho):
  from IPython.display import HTML
  from base64 import b64encode
  mp4 = open(video_caminho, 'rb').read()
  data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
  return HTML("""
  <video width=500 controls>
    <source src="%s" type="video/mp4">
  </video>
  """ % data_url)
```
